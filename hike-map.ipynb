### Import modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import folium
from folium import IFrame
import base64
from PIL import Image
from io import BytesIO
```

### Create hike dataframe

For this to work properly, all I need to start is coordinate data and a code to connect the hike and its photo. Other features will come later.

```python
df = {'lat': [45.229, 45.402, 37.624],
      'lon': [-117.274, -121.776,-112.16],
      'code': ['bald_mountain', 'ice_lake', 'queens_garden'],
      'name': ['Bald Mountain, Mt. Hood, OR', 'Ice Lake, Joseph, OR', 
               'Queens Garden, Bryce Canyon National Park, UT'],
      'month': [7, 7, 5],
      'year': [2022, 2022, 2022]
     }

data = pd.DataFrame(df)
data
```

### Initialize map

```python
world = folium.Map(zoom_start=4, location=[42, -115])
            
x = data[['lat', 'lon', 'code', 'name']].copy()

for index, row in x.iterrows():
    code = row['code']
    name = row['name']
    encoded = base64.b64encode(open(f'photos/{code}.png', 'rb').read())
    html = '<img src="data:image/png;base64,{}">'.format
    iframe = IFrame(html(encoded.decode('UTF-8')), width=500, height=400)
    popup = folium.Popup(iframe, max_width=500) 
    folium.Marker([row['lat'], row['lon']], tooltip=name, popup=popup, 
              icon=folium.Icon(color = 'gray')).add_to(world)               
              
world
```
